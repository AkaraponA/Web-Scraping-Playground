```python
from bs4 import BeautifulSoup
import requests
```

```python
resource = requests.get('https://stackpython.co/tutorials')
```

```python
if resource.status_code == 200:
    print("Successful")
elif resource.status_code == 404:
    print("Error 404 Page not found")
else:
    print("Not both 200 and 404")
```

```python
soup = BeautifulSoup(resource.text, 'html.parser')
print(soup.prettify())
```

```python
soup.title
```

```python
tutorial_title = soup.find_all('a', class_="card-title")
```

```python
python_title = []
for title in tutorial_title:
    print(title.text, end='\n'*2)
    python_title.append(title.text)
```

```python
python_title
```

```python
tutorial_date = soup.find_all('p', class_="card-text")
```

```python
story_date = []
for date in tutorial_date:
    print(date.text, end='\n'*2)
    story_date.append(date.text)
```

```python
story_date
```

```python
import pandas as pd
```

```python
stackpython = pd.DataFrame(zip(python_title,story_date),columns=['Tutorial Name','Tutorial Create'])
```

```python
stackpython
```

```python
stackpython.to_csv('Stackpython tutorial list.csv',encoding='utf-8')
```
